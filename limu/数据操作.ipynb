创建一个元素全为 0 的张量（Tensor），括号中的参数表示张量的维度大小，每个数字对应一个维度的长度。

第一个参数 2：表示张量的第 0 维（最外层维度） 大小为 2，即该维度有 2 个元素。

第二个参数 3：表示张量的第 1 维大小为 3，即每个外层元素包含 3 个中层元素。

第三个参数 4：表示张量的第 2 维（最内层维度） 大小为 4，即每个中层元素包含 4 个最内层元素。

因此，torch.zeros(2, 3, 4) 会创建一个形状为 (2, 3, 4) 的三维张量，总元素数量为 2×3×4 = 24 个（均为 0）。

可以理解为：这个张量包含 2 个 "块"[ , ]；每个块又分3个块，即上一个[ , ]中的空里是[ , , ]；每个块再分4个块，即上一个[ , ]中的空里是[ , , , ]

```python
x = torch.arange(12,dtype=torch.float32).reshape((3,4))# 形状为(3,4)
x
```

```python
y = torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])# 形状为(3,4)
y
```

```python
torch.cat((x,y),dim=0)# 第0维是最外围，拼完后形状变成(6,4)
```

```python
torch.cat((x,y),dim=1)# 拼完后形状变成(3,8)
```

torch.cat((x, y), dim=0) 是用于拼接张量的函数，其参数含义如下：

第一个参数 (x, y)：这是一个包含待拼接张量的元组（或列表），表示需要将 x 和 y 这两个张量按照指定维度拼接起来。

注意：x 和 y 必须满足 “除拼接维度外，其他所有维度的大小都相同”，否则会报错。

第二个参数 dim=0：表示拼接的维度，dim=0 即沿着第0维进行拼接。维度的编号从0开始。

即使形状不同，仍然可以通过调用广播机制俩执行按元素操作

```python
a = torch.arange(3).reshape((3,1))
b = torch.arange(2).reshape((1,2))
a,b
```

```python
a+b #维度不同，由于广播机制，使得a被复制成3x2的矩阵，b被复制成3x2的矩阵，再相加
```

```python
before = id(y)# id告诉我们对象y在python中的标识号，类似c++的指针
y = y+x
id(y) == before
```

```python
z = torch.zeros_like(y)# 生成一个与输入张量y形状完全相同、但所有元素值均为0的新张量
print('id(z):',id(z))
z[:] = x+y# z中所有元素=x+y中的元素
print('id(z):',id(z))
```

```python
#如果后续运算中不重复使用x，可以使用x[:]=x+y或x+=y减少操作内存开销
before = id(x)
x += y
id(x) == before
```
