```python
import os
```

```python
os.makedirs(os.path.join('..','data'),exist_ok=True)# 创建目录
# os.path.join('..','data')拼接出上级目录下的data文件夹路径
# os.makedirs(..., exist_ok=True)创建这个文件夹，exist_ok=True表示如果文件夹已存在也不会报错

data_file = os.path.join('..','data','house_tiny.csv')# 定义文件路径
# 定义了要创建的CSV文件路径：上级目录的data文件夹下的house_tiny.csv

# 写入数据
with open(data_file,'w') as f:# 以写入模式（'w'）打开文件
  # 第一行写入CSV表头：NumRooms（房间数量）、Alley（小巷类型）、Price（价格）
  f.write('NumRooms,Alley,Price\n')
  # 后续几行写入具体数据，其中NA表示缺失值
  f.write('NA,Pave,127500\n')
  f.write('2,NA,10600\n')
  f.write('4,NA,178100\n')
  f.write('NA,NA,140000\n')
```

```python
import pandas as pd

data = pd.read_csv(data_file)
print(data)
data
```

```python
inputs, outputs = data.iloc[:,0:2], data.iloc[:,2]
inputs = inputs.fillna(inputs.mean())
print(inputs)
```

```python
inputs, outputs = data.iloc[:,0:2], data.iloc[:,2]
inputs1 = data.iloc[:,0]
data.iloc[:,0] = data.iloc[:,0].fillna(inputs1.mean())# 用均值填充NumRooms列的NAN
print(inputs)
inputs = pd.get_dummies(inputs,dummy_na=True,dtype=int)# 若为False,则没有Alley_nan列
print(inputs)
```

Alley_nan 列的出现是 dummy_na=True 参数的预期效果，它的作用是明确标记原始数据中哪些样本的 Alley 列存在缺失值，避免缺失信息被忽略。如果不想要这一列，可以将参数改为 dummy_na=False（默认值），此时缺失值会被直接忽略，只生成非缺失类别的哑变量列。
